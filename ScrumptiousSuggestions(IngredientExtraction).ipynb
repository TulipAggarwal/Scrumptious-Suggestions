```python
import pandas as pd
file_path = '/content/IndianFoodDataset(english)modified.xlsx'
df = pd.read_excel(file_path)

# Spliting the values in the "Ingredients" column where there are multiple values in a cell
ingredient_series = df['Ingredients'].str.split(',').explode()

# Extracting unique values from the "Ingredients" column, drop NaN values, and sort them in ascending order
unique_ingredients_df = pd.DataFrame({'Ingredients': ingredient_series.unique()})
unique_ingredients_df = unique_ingredients_df.dropna().drop_duplicates().sort_values(by='Ingredients', ascending=True).reset_index(drop=True)

# Removing empty values, signs, and specific words from the "Ingredients" column
unique_ingredients_df['Ingredients'] = (
    unique_ingredients_df['Ingredients']
    .replace(r'^\s*$', '', regex=True)
    .replace(r'^[^\w\s]+', '', regex=True)
    .replace(r'\b(cup|cups|tablespoon|tablespoons|teaspoon|teaspoons|liter|Liter|Long|long|Kg|kg|Veeba|veeba)\b', '', regex=True)
    .replace(r'[^\w\s]', '', regex=True)
    .str.strip()
    .str.title()
)

# Removing rows with empty "Ingredients" values
unique_ingredients_df = unique_ingredients_df[unique_ingredients_df['Ingredients'] != '']

# Sorting the DataFrame alphabetically
unique_ingredients_df = unique_ingredients_df.sort_values(by='Ingredients', ascending=True).reset_index(drop=True)
unique_ingredients_df = unique_ingredients_df.drop_duplicates()

# Saving the cleaned, sorted, and capitalized DataFrame to a new Excel file named 'cleaned_sorted_capitalized_ingredients.xlsx'
output_file_path = 'Ingredients.xlsx'
unique_ingredients_df.to_excel(output_file_path, index=False)

unique_ingredients_df.head(50)
```
