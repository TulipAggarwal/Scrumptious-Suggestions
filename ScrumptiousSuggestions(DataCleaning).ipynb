#Making separate dataframes for Hindi and English recipes

```python
from langdetect import detect

def detect_language(text):
    try:
        return detect(str(text))
    except:
        return 'unknown'

# Applying the language detection function to each row
df['DetectedLanguage'] = df['Instructions'].apply(detect_language)

# Creating separate DataFrames for Hindi and English
df_hindi = df[df['DetectedLanguage'] == 'hi']
df_english = df[df['DetectedLanguage'] == 'en']

# Display or save the DataFrames
df_hindi.head(10)
```

```python
df_hindi.to_excel('IndianFoodDataset(hindi).xlsx', index=False)
df_english.to_excel('IndianFoodDataset(english).xlsx', index=False)
```
